
## Brightness calculation, general formula

The brightness of the undulator source $B$ is defined as:   
(Eq. 2.1, _**Elements of Modern X-ray Physics**_, Als-Nielsen and McMorrow, 2nd Edition)
$$
B = \frac{F\ (\text{photons}/\text{sec})}{A_{d}\ (\text{mrad}^2)\cdot S_{s}\ (\text{mm}^2) \cdot E_{0.1\% bw}}
$$
Here:

$F$ is the x-ray flux in unit of $\text{photons/second}$;  

$S_{s}= 2\pi\sigma_x\sigma_y = 6.99 \times 10^{-4}\text{ mm}^2 $ is the 2D Gaussian electron beam size measured at 3-ID. Here $\sigma_x$ and $\sigma_y$ are the standard deviation of the electron beam;

$A_{d} = (S_xS_y)/D^{2}$ is the divergence of the slit in unit of $\text{mrad}^2$. Here $S_x$ and $S_y$ are the sizes of the slit and $D$ is the distance between the slit and the undulator source;  

$E_{0.1\% bw}$ is 0.14 for a Si (111) monochromator and does not vary significantly as a function of energy for thick crystals.


## Brightness calculation at 8-ID

Flux is measured using Nino's calibrated pin diode (DP00429) mounted at the 8-ID-E diffractometer. The Spectral Responsivity at 8 keV is 0.2637 $\text{W/A}$.

The beam is slited down to 100 $\mu m$ x 100 $\mu m$ at a distance 54600 mm downstream of the undulator source (slit 5).


```python

Pind_Current = 250.9e-6                
Photon_Energy = 8e3                      # Photon energy in eV
Spec_Res = 0.2637
loss_mirror = 0.94*0.94                  # Loss from mirror reflectivity
loss_diamond = 0.735                     # Loss from BPM and CVD diamond window at 8-ID-E
norm_ring_current = 1.0                  
loss_factor = loss_mirror*loss_diamond*norm_ring_current

# Constants for flux calculation
ev2J = 1.6e-19                           # Conversion from eV to Joule  
E_0p1bw = 9.89e-05/1e-3                  # From Xianbo's calculation on 07/22/2025  

# Calculation of raw and corrected flux
Beam_Power = Pind_Current/Spec_Res
Raw_Flux = Beam_Power/(Photon_Energy*ev2J)       # Flux in photons per second
Corrected_Flux = Raw_Flux/loss_factor

# Constants for brightness calculation
S_s = 6.99e-4
D = 54600                                # Distance from the source to sl5 in mm
S_x = 0.1                                # Size of sl5 horizontal in mm
S_y = 0.1                                # Size of sl5 vertical in mm

# Calculation of aperture divergence
A_d = (S_x/D*1e3)*(S_y/D*1e3)  

# Calculation of brightness
B = Corrected_Flux/(S_s*A_d*E_0p1bw)
```

```python

print(f"\n The measured raw flux of the 8-ID-I beamline at {Photon_Energy/1e3:.1f} keV is: {Raw_Flux:.2e} \
photons/sec")

print(f"\n The total loss factor is: {loss_factor:.3f}")

print(f"\n The corrected flux of the 8-ID-I beamline at {Photon_Energy/1e3:.1f} keV is: {Corrected_Flux:.2e} \
photons/sec")

print(f"\n The brightness of the 8-ID-I beamline is: {B:.2e} \
photons/sec/mrad²/mm²/0.1%bw ")
```
