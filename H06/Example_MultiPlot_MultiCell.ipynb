```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
```

```python

#Example with 2 rows and 2 columns.
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
```

```python

# Cell 1: axes[0, 0]
axes[0,0].plot(x,y1)
axes[0,0].set_title("sin(x)")
plt.figure(fig)
plt.show()
```

```python

# Cell 2: axes[0, 1]
axes[0,1].plot(x,y2)
axes[0,1].set_title("cos(x)")
plt.figure(fig)
plt.show()
```

```python


# Cell 3: axes[1, 0]
axes[1,0].plot(x,y3)
axes[1,0].set_title("x^2")
plt.figure(fig)
plt.show()
```

```python

# Cell 4: axes[1, 1]
axes[1,1].plot(x, x)
axes[1,1].set_title("y=x")
plt.figure(fig)
plt.show()
```

```python


#Cell 5: tight layout and show all
plt.figure(fig)
plt.tight_layout()
plt.show()

```
