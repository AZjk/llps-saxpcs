```python
def get_temperature_and_g2(avg_data_list, target_idx=44, g2_idx=0):
    xy = []
    for avg_dict in avg_data_list:
        t = np.nanmean(avg_dict['temperature'])
        g2 = avg_dict['g2'][target_idx][g2_idx]
        g2_err = avg_dict['g2_err'][target_idx][g2_idx]
        xy.append((t, g2, g2_err))
    xy = np.array(xy)
    return xy
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2.4, 2))
labels = ['H02', 'H04', 'H05', 'H06']

import itertools
marker = itertools.cycle(('+', '.', 's', '*')) 

for n, avg_data_list in enumerate([avg_H02, avg_H04, avg_H05, avg_H06]):
    t_g2 = get_temperature_and_g2(avg_data_list)
    # ax.plot(t_g2[:, 0], t_g2[:, 1], marker=next(marker), label=labels[n])
    ax.errorbar(t_g2[:, 0], t_g2[:, 1], yerr=t_g2[:, 2], marker=next(marker), label=labels[n], ms=3)

# labels = ['H06-100%', 'H06-50%', 'H06-25%']
# for n, data_dict in enumerate([avg_H06_100, avg_H06_050, avg_H06_025]):
#     ax.plot(data_dict['temperature'], data_dict['g2'][:, 44, 0], lw=0.5, ms=1, label=labels[n])

ax.set_xlabel('Temperature ($^\circ$C)')
ax.set_ylabel('g2(dt=0.1s)')
ax.set_ylim(0.95, 1.25)
plt.legend()
plt.subplots_adjust(wspace=0)
plt.savefig('figure4.png', dpi=300)
# plt.savefig('figure4.pdf')
plt.tight_layout()
```
