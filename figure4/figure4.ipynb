```python
def get_temperature_and_g2(avg_data_list, target_idx=44):
    xy = []
    for avg_dict in avg_data_list:
        t = np.nanmean(avg_dict['temperature'])
        g2 = avg_dict['g2'][target_idx][0]
        xy.append((t, g2))
    xy = np.array(xy)
    return xy
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3.8, 4))
labels = ['H02', 'H04', 'H05', 'H06-100%', 'H06-50%', 'H06-25%']

import itertools
marker = itertools.cycle(('+', '.', 'o', '*')) 

for n, avg_data_list in enumerate([avg_H02, avg_H04, avg_H05, avg_H06_100, avg_H06_050, avg_H06_025]):
    t_g2 = get_temperature_and_g2(avg_data_list)
    ax.plot(t_g2[:, 0], t_g2[:, 1], marker=next(marker), label=labels[n])

plt.legend()
plt.subplots_adjust(wspace=0)
plt.savefig('figure4.png', dpi=300)
plt.savefig('figure4.pdf')
plt.tight_layout()
```
